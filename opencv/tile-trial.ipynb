```python
# IMPORTS:
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import openslide
```

```python
def slide_details (slide, showThumbnail = True, maxSize = (600, 400)):
    if showThumbnail:
        display(slide.get_thumbnail(size = maxSize))
    
    print(f"File id: {slide}")
    print(f"Dimensions: {slide.dimensions}")
    print(f"Number of Levels: {slide.level_count}")
    print(f"Downsample per Level: {slide.level_downsamples}")
    print(f"Level Dimensions: {slide.level_dimensions}")
```

```python
# LOAD AND DISPLAY
wsi = openslide.OpenSlide(r'E:\WSIs\259249.svs')
slide_details(wsi)
wsi.close()
```
